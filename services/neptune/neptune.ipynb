# person 頂点（例：Alice）

```python
%%gremlin
g.addV('person')
 .property('id', 'p1')
 .property('name', 'Alice')
```

# city 頂点（例：Tokyo)

```python
%%gremlin
g.addV('city')
 .property('id', 'c1')
 .property('name', 'Tokyo')
```

# 登録した頂点を確認

```python
%%gremlin
g.V().valueMap(true)
```

# エッジを登録

```python
%%gremlin
g.V().has('person','name','Alice').as('p')
 .V().has('city','name','Tokyo').as('c')
 .addE('lives_in').from('p').to('c')
```

# 登録したエッジを確認

```python
%%gremlin
g.E().valueMap(true)
```

```python
%%gremlin
g.V().has('person','name','Alice').outE('lives_in').inV()
```
